```python
import matplotlib.pyplot as plt
from matplotlib.offsetbox import TextArea

# Create the figure and axes
fig, ax = plt.subplots()

# Text content with line breaks for illustration (replace with your text)
text = "This is a text with multiple lines\nand it will wrap automatically"

# Create a TextArea object with wrapping enabled
text_box = TextArea(text, textprops=dict(wrap=True))

# Get the bounding box of the axes (normalized coordinates)
bbox = ax.bbox

# Calculate desired position (adjust as needed)
x_pos = 0.7  # Adjust for horizontal position (0 to 1)
y_pos = 0.8  # Adjust for vertical position (0 to 1)

# Use figtext to display the text box at the calculated position
plt.figtext(x_pos, y_pos, text_box, ha="center", va="center",color='k')

# Display the plot
plt.show()
plt.savefig('output1.png')
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Arial'

fig = plt.figure(dpi=300)
plt.axis('off')
plt.axis([0, 10, 0, 5])
#plt.ylim(2,5)

loc, addr = ['Kolkata, West Bengal, India', '27C, S Sinthee Rd, Pearabagan, South Sinthee, Biswanath Colony, Sinthee, Kolkata, West Bengal 700050, India\nLat\nLon\nDate']

plt.text(0, 4, loc, fontsize=11.5, ha='left', va='top', wrap=True, color='white', weight='semibold')
plt.text(0, 3.7, addr, fontsize=8.5, ha='left', va='top', wrap=True, color='white', weight='semibold',linespacing=1.7)

plt.gcf().set_facecolor('none')

# Save the plot with transparent background
plt.savefig('output.png', transparent=True, dpi=300)
```

```python
def paste_img_on_img(background, foreground, shiftL=0, shiftD=0):
    # Get the dimensions of both images
    bg_width, bg_height = background.size
    fg_width, fg_height = foreground.size

    # Calculate the bottom right corner coordinates for placement
    right = bg_width - fg_width - shiftL  # Places at right edge, adjust if needed
    bottom = shiftD  # Places at bottom edge

    # Paste the small image onto the large image
    background.paste(foreground, (right, bottom), foreground)

    return background
```

```python
from PIL import Image
```

```python
f = Image.open('./NEW_test.png')
t = Image.open('./output.png')
width, height = t.size
t = t.resize((int(width*0.7),int(height*0.7)))

ft = paste_img_on_img(f,t,shiftL=90,shiftD=670)
ft.save('super_Test.png')
```
